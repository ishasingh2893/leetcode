```python
# Create an empty list & append new entries
names = []
ages = []
emails = []

#iterating through a loop
for i in range(3):
    print(i + 1, 'Input')
    name = input('Name: ')
    age = input('Age: ')
    email = input('Email: ')

    names.append(name) #notice the append function
    ages.append(age)
    emails.append(email)

print("Names:", names)
print("Ages:", ages)
print("Emails:", emails)
```

```python
# Instead of scouring through 3 different lists for the info of one person, we can have one consolidated dictionary
contacts = [] #creating an empty list

for i in range(3):
    print(i + 1, 'Input')
    name = input('Name: ')
    age = input('Age: ')
    email = input('Email: ')

    #create a dictionary
    person = { 
        'name': name,
        'age': age,
        'email': email
    }

    # append that dictionary to our empty list
    contacts.append(person)
print(contacts)
    
```

```python
# Add/delete/search contacts

def add_contact():
    name = input('Name: ')
    age = input('Age: ')
    email = input('Email: ')
    person = { 
        'name': name,
        'age': age,
        'email': email
    }
    return person

def display_people(contacts):
    for i, person in enumerate(contacts):
        print(i+1,'-', person['name'], '|', person['age'], '|',person['email'])
        
def delete_contact(contacts):
    display_people(contacts)
    
    while True:
        number = input('Enter a number to delete: ')
        try:
            number = int(number)
            if number <=0 or number > len(contacts):
                print('Invalid number. Out-of-range')
            else:
                break
        except:
            print('Invalid number')
        
    contacts.pop(number-1)
    print ('Contact deleted!')


def search(contacts):
    search_name = input('Search for a name: ').lower()
    results = []

    for person in contacts:
        name = person['name']
        if search_name in name.lower():
            results.append(person)
    display_people(results)
        
    
    
print ('Hi, welcome to the contact management system!')

contacts = [
    {'name': 'Tim', 'age': 24, 'email': 'tim@gmail.com'},
    {'name': 'Dim', 'age': 34, 'email': 'dim@gmail.com'},
    {'name': 'Lim', 'age': 64, 'email': 'lim@gmail.com'},
    {'name': 'Timothy', 'age': 54, 'email': 'lim@gmail.com'},
]

while True:
    print ('Contact list size :', len(contacts))
    command = input ('What do you wanna do? (Enter add/delete/search/or, Q for quit): ').lower()
    
    if command == 'add':
        person = add_contact()
        contacts.append(person)
        print('Contact added!')
    elif command == 'delete':
        delete_contact(contacts)
    elif command == 'search':
        search(contacts)
    elif command == 'q':
        break
    else:
        print('Invalid command')

print(contacts)
```
