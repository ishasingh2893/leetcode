```python
def print_board(board):
    for i, row in enumerate(board):
        row_str = ' '
        for j, value in enumerate(row):
            row_str += value
            if j != len(row) - 1:
                row_str += ' | '
        print (row_str)
        if i != len(board) - 1:
            print('-----------')

def get_move(turn, board):
    #ask the user for a row/col where they wanna enter their choice
    while True:
        row = int(input('Row: '))
        col = int(input('Col: '))

        if row < 1 or row > len(board):
            print('Invalid row, try again.')
        elif col < 1 or col > len (board[row - 1]):
            print('Invalid col, try again.')
        elif board[row - 1][col - 1] != ' ':
            print('Already taken, try again')
        else:
            break
    board[row - 1][col - 1] = turn

def check_win(board, turn):
    lines = [
        [(0,0), (0,1), (0,2)],
        [(1,0), (1,1), (1,2)],
        [(2,0), (2,1), (2,2)],
        [(0,0), (1,0), (2,0)],
        [(0,1), (1,1), (2,1)],
        [(0,2), (1,2), (2,2)],
        [(0,0), (1,1), (2,2)],
        [(0,2), (1,1), (2,0)],
    ]

    for line in lines:
        win = True
        for position in line:
            row, col = position
            if board[row][col] != turn:
                win = False
                break
        if win:
            return True
    return False

#list within a list/ multi-dimensional array
# board [0][2] will give us the element of first list which is at the second index position

board = [
    [' ', ' ', ' '],
    [' ', ' ', ' '],
    [' ', ' ', ' '],
]

turn = 'X'
turn_number = 0
print_board (board)

while turn_number <= 9:
    print('It is the', turn, 'players turn. Please make the move.')
    
    get_move(turn, board)
    
    print_board(board)
    
    winner = check_win(board, turn)
    
    if winner:
        break
    if turn == 'X':
        turn = 'O'
    else:
        turn = 'X'
    turn_number += 1

if turn_number == 9:
    print('Game tied!')
else:
    print ('The winner is', turn)
```
