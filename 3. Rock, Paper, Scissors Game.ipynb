```python
import random
```

```python
user_wins = 0
computer_wins = 0
```

```python
options = ['rock', 'paper', 'scissors']
```

```python
while True:
    user_input = input('Type Rock/Paper/Scissors or Q to quit: ').lower()
    if user_input == 'q':
        break # this will break us out of the while-loop and take us to the end message of goodbye
   
    if user_input not in options: # if the user typed none of these 3
        continue # instead of an error message, this will display the initial prompt (Type R/P/S); nothing beyond this point will be executed

    random_number = random.randint(0, 2) # rock: 0, paper:1, scissors:2
    computer_pick = options[random_number]
    print('Computer picked', computer_pick + ".")

    # Game conditions
    if user_input == 'rock' and computer_pick == 'scissors':
        print('You won!')
        user_wins += 1

    elif user_input == 'paper' and computer_pick == 'rock':
        print('You won!')
        user_wins += 1
      
    elif user_input == 'scissors' and computer_pick == 'paper':
        print('You won!')
        user_wins += 1
        
    else:
        print('You lost!')
        computer_wins += 1
    
print('You won', user_wins, 'times.')
print('The computer won', computer_wins, 'times.')
print('Goodbye!')
```
