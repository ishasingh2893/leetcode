```python
playing = input("Do you want to play? ").lower()
```

```python
print (playing)
```

```python
if playing.lower() != 'yes':
    quit()
print("Okay! Let's play:)")
score = 0
    
''' This converts the user's response to all lowercase and if they entered 'yes' in any format, 
then the code will work. Can also use .upper() but change the input value according.

We can also use these two functions like this:
1) answer = input('What does CPU stand for? ').lower() , or
2) if answer.lower() == 'central processing unit':\

'''
```

```python
answer = input('What does CPU stand for? ').lower()
if answer == 'central processing unit':
    print('Correct!')
    score += 1
else:
    print('Incorrect!')
```

```python
answer = input('What does GPU stand for? ').lower()
if answer == 'graphics processing unit':
    print('Correct!')
    score += 1
else:
    print('Incorrect!')
```

```python
answer = input('What does RAM stand for? ').lower()
if answer == 'random access memory':
    print('Correct!')
    score += 1
else:
    print('Incorrect!')
```

```python
answer = input('What does PSU stand for? ').lower()
if answer == 'power supply unit':
    print('Correct!')
    score += 1
else:
    print('Incorrect!')

print("You got " + str(score) + " questions correct!")
print("You got " + str((score/4) * 100) + "%")
```
