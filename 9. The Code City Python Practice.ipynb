Convert C to F

```python
def temp_converter(celsius):
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit

while True:
    temperature = input("Enter temperature in °C or Q to quit: ").lower()

    if temperature == 'q':
        print("Thanks for playing!")
        break

    try:
        temperature = float(temperature)
        converted_temp = temp_converter(temperature)
        print(f"{temperature}°C = {converted_temp}°F")

    except ValueError:
        print("Please enter a valid number or Q to quit.")
```

Drawing pyramids

```python
rows = 10

for i in range(1, rows + 1):
    # print spaces
    print(' ' * (rows-i), end = '')
    # print asterisks
    print('*' * (2*i - 1))
```

```python
rows = 10

for i in range(1, rows + 1):
    stars = '*' * (2 * i - 1)
    print(stars.center(2 * rows - 1))
```

```python
rows = 10

for i in range(1, rows+1):
    print(('★' * (2*i - 1)).center(2*rows - 1))
```

Prime number check

```python
number = int(input("Enter a number: "))

if number <= 1:
    print("Not prime")

else:
    is_prime = True # created a variable and stored a boolean value (Assumption)

    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            is_prime = False # break assumption, update the variable
            break

    if is_prime:
        print("Prime number")
    else:
        print("Not prime")
```

```python
# make a list of factors

number = int(input("Enter a number: "))

factors = []

for i in range(1, number + 1):
    if number % i == 0:
        factors.append(i)

print("Factors:", factors)

if len(factors) == 2:
    print("Prime number")
else:
    print("Not prime")
```

```python
# another method

number = int(input("Enter a number: "))

factors = []

for i in range(1, int(number ** 0.5) + 1):

    if number % i == 0:

        factors.append(i)

        if i != number // i:
            factors.append(number // i)

factors.sort()

print("Factors:", factors)

if len(factors) == 2:
    print("Prime number")
else:
    print("Not prime")
```

Palindrome check

```python
# Reverse == Original

def is_palindrome(org_string):
    rev_string = org_string[: : -1]
    return org_string == rev_string

org_string = input('Enter your word: ').lower()
print(is_palindrome(org_string))
    
```

```python
def is_palindrome(org_string):
    return org_string == org_string[::-1]

org_string = input('Enter your word: ').lower()
print(is_palindrome(org_string))
```

```python
def is_palindrome(text):
    text = text.lower()
    return text == text[::-1]

text = input("Enter your word: ")

print(is_palindrome(text))
```
