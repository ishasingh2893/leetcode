```python
operand_1 = input('Number 1: ')
operand_2 = input('Number 2: ')
sign = input('Sign: ')

try:
    operand_1 = float(operand_1) # converting the string inputs to floats 
    operand_2 = float(operand_2)
except:
    print('Invalid operand')

result = 0 # to handle the division by zero case
if sign == '+':
    result = (operand_1) + (operand_2)
elif sign == '-':
    result = (operand_1) - (operand_2)
elif sign == '/':
    if float(operand_2)!= 0:
        result = (operand_1) / (operand_2)
    else:
        print('Division by zero.') 
elif sign == '*':
    result = (operand_1) * (operand_2)
print(result)
```
