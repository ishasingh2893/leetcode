Method I

```python
operand_1 = input('Number 1: ')
operand_2 = input('Number 2: ')
sign = input('Sign: ')

# converting the string inputs to floats and preventing the code from failing if anything other than a number has been entered 
try:
    operand_1 = float(operand_1) 
    operand_2 = float(operand_2)
    result = 0 # to handle the division by zero case
    if sign == '+':
        result = (operand_1) + (operand_2)
    elif sign == '-':
        result = (operand_1) - (operand_2)
    elif sign == '/':
        if float(operand_2)!= 0:
            result = (operand_1) / (operand_2)
        else:
            print('Division by zero.') 
    elif sign == '*':
        result = (operand_1) * (operand_2)
    print(result)

except:
    print('Invalid operand')
```

Method III

```python
# Recognizing an invalid input and asking the user to enter the correct input

while True:
    operand_1 = input('Number 1: ')
    try:
        operand_1 = float(operand_1)
        break
    except:
        print ('Invalid input number 1. Please enter a number.')

while True:
    operand_2 = input('Number 2: ')
    try:
        operand_2 = float(operand_2)
        break
    except:
        print ('Invalid input number 2. Please enter a number.')

sign = input('Sign: ')

result = 0 # to handle the division by zero case
if sign == '+':
    result = (operand_1) + (operand_2)
elif sign == '-':
    result = (operand_1) - (operand_2)
elif sign == '/':
    if float(operand_2)!= 0:
        result = (operand_1) / (operand_2)
    else:
        print('Division by zero.') 
elif sign == '*':
    result = (operand_1) * (operand_2)
print(result)
```

Method IV

```python
# instead of writing two while-loops, we can write a function

def get_number(number): #parameter
    while True:
        operand = input('Number ' + str(number) + ':') # Telling the user to input the number
        try:
            return float(operand)
        except:
            print ('Invalid input. Please enter a number.')

operand_1 = get_number(1) #argument 
operand_2 = get_number(2)
sign = input('Sign: ')

result = 0 # to handle the division by zero case
if sign == '+':
    result = (operand_1) + (operand_2)
elif sign == '-':
    result = (operand_1) - (operand_2)
elif sign == '/':
    if float(operand_2)!= 0:
        result = (operand_1) / (operand_2)
    else:
        print('Division by zero.') 
elif sign == '*':
    result = (operand_1) * (operand_2)
print(result)
```
