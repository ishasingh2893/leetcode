Write a random number generating function that asks the user to enter a number > then,
check if that is actually a number greater than 0 (and not a string or negative number > then,
convert that input into an integer.

```python
top_of_range = input('Type a number: ')

if top_of_range.isdigit():
    top_of_range = int(top_of_range)

    if top_of_range <= 0:
        print('Please type a number greater than 0.')
        quit()
else:
    print('Please type a number.')
    quit()
```

Using the randint function, generate a random number less than the user's input and ask the user to make a guess, based on which, the function provides a feedback.

```python
import random
```

```python
random_number = random.randint(0, top_of_range) 
# here, we need to give the bottom limit as well and not just the variable name. Hence, 0 included.
```

```python
print(random_number)
```

```python
while True:
    user_guess = input('Make a guess: ')
    if user_guess.isdigit():
        user_guess = int(user_guess)
    else:
        print('Please type a number.')
        continue

    if user_guess == random_number:
        print('You got it!')
        break
    else:
        print('You got it wrong! :( ')
```

Notice that we can also put a 'break' after the user enters a wrong guess but then the function will get out of the loop immediately, no matter if the user was right or wrong. Putting a break only after the correct answer has been guessed keeps letting the user to guess.
